```python
import numpy as np
from sklearn.cluster import KMeans
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Racoon = scipy.misc.face(gray=True)
plt.figure(figsize=(10,5))
plt.imshow(Racoon,plt.cm.gray)
plt.show()
```

```python
rows = Racoon.shape[0]
columns = Racoon.shape[1]
image = Racoon.reshape(rows*columns,1)
```

## Kmeans clustering to fit the image:

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(image)
```

```python
clusters = np.asarray(kmeans.cluster_centers_)
labels = np.asarray(kmeans.labels_)
labels = labels.reshape(rows,columns)
```

```python
labels.shape
```

```python
# Saving the compressed image:
plt.imsave('compressed.png',labels)
```

```python
image = plt.imread('compressed.png')
plt.figure(figsize=(12,6))
plt.imshow(image)
plt.show()
```

## Noticeable quality difference using kmeans with 5 clusters.
